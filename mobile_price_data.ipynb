# Analysis of mobile_price_data

## Information about dataset attributes :-

* **mobile_name** : The name or model of the mobile phone.
* **mobile_price** : The price of the mobile phone.
* **mobile_color** : The color of the mobile phone.
* **dual_sim** : Indicates whether the phone supports dual SIM cards.
* **disp_size** : The display size of the phone in inches.
* **resolution** : The screen resolution of the phone.
* **os** : The operating system used by the phone.
* **num_cores** : The number of CPU cores in the phone's processor.
* **mp_speed** : The speed of the phone's processor in GHz.
* **int_memory** : The internal storage capacity of the phone in gigabytes.
* **ram** : The random access memory (RAM) capacity of the phone in gigabytes.
* **p_cam** : The resolution of the rear camera.
* **f_cam** : The resolution of the front-facing camera.
* **network** : The type of network supported by the phone (e.g., 4G, 5G).
* **bluetooth** : Indicates whether the phone supports Bluetooth technology.
* **battery_power** : The battery power or capacity of the phone in milliampere-hours (mAh).
* **mob_width** : The width of the phone.
* **mob_height** : The height of the phone.
* **mob_depth** : The depth or thickness of the phone.
* **mob_weight** : The weight of the phone.

## Encoding Data

```python
data = data.drop(['mobile_name' ,'mobile_color','dual_sim'], axis = 1)
```

```python
object=['p_cam' ,'f_cam' ,'network', 'disp_size','resolution','os',	'num_cores']

from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
for cols in object:
    data[cols] = encoder.fit_transform(data[[cols]])
```

```python
data.dtypes
```

```python
data.sample(5)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split  
X = data.drop('mobile_price' ,axis=1)  
y = data['mobile_price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)  
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 25, criterion = 'entropy', random_state =0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
a=accuracy_score(y_test, y_pred) 
print(a)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
classifier1 = DecisionTreeClassifier( criterion = 'entropy', random_state = 0 ,max_depth=10)
classifier1.fit(X_train, y_train)
y_pred1 = classifier1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
b=accuracy_score(y_test, y_pred1) 
print(b)
```

```python
cm1 = confusion_matrix(y_test, y_pred1)
print(cm1)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
classifier2 = GaussianNB()
classifier2.fit(X_train, y_train)
y_pred2 = classifier2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
c=accuracy_score(y_test, y_pred2) 
print(c)
```

# K-Nearest Neighbors (K-NN)

```python
from sklearn.neighbors import KNeighborsClassifier
classifier3 = KNeighborsClassifier(n_neighbors = 1, metric = 'minkowski', p = 20)
classifier3.fit(X_train, y_train)
y_pred3 = classifier3.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
d=accuracy_score(y_test, y_pred3) 
print(d)
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
classifier4 = SVC(kernel = 'linear', random_state = 0)
classifier4.fit(X_train, y_train)
y_pred4 = classifier4.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
e=accuracy_score(y_test, y_pred4) 
print(e)
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression  
classifier5= LogisticRegression(random_state = 0 )  
classifier5.fit(X_train, y_train) 
y_pred5=classifier5.predict(X_test)   
   
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix  
f =accuracy_score(y_test, y_pred5)   
print(f)
```
